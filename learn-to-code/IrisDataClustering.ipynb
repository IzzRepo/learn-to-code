Prior to K Means clustering, we need to do PCA

Feature scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
features = ['sepal length (cm)',' sepal width (cm)','petal length (cm)','petal width (cm)']

# convert from dataframe to numpy array
x = df.iloc[:,:].values
```

```python
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2)
```

```python
principalComponents = pca.fit_transform(x)
```

```python
df_pca = pd.DataFrame(principalComponents, columns = ['Feature1','Feature2'])
```

```python
df_pca.head()
```

K-Means

```python
X = df_pca.iloc[:,:].values
```

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
```

```python
model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,12))
```

```python
visualizer.fit(X)
visualizer.show()
```

```python
# using elbow method to determine K values for KMeans clustering
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# apply kmeans to dataset
kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 0)
y_kmeans = kmeans.fit_predict(X)
```

```python
# Visualize the clusters
plt.scatter(x[y_kmeans == 0,0], x[y_kmeans == 0,1], s = 100, c = 'red', label = 'C1')
plt.scatter(x[y_kmeans == 1,0], x[y_kmeans == 1,1], s = 100, c = 'blue', label = 'C2')
plt.scatter(x[y_kmeans == 2,0], x[y_kmeans == 2,1], s = 100, c = 'green', label = 'C3')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s = 300, c = 'yellow', label = 'Centroids')
plt.title("Clusters of Iris data")
plt.xlabel("feature1")
plt.ylabel("feature2")
plt.legend()
plt.show()
```
